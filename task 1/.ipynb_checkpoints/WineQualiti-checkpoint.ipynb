## Обычная Линейная Нейросеть 
Нейросеть выдаёт вероятности принадлежности к каждому классу, затем пропускается через экспоненты, для многократного уменьшения мантисы числа, и берётся вырез значения из списка классов.

```python
class simple_predict(nn.Module):
    def __init__(self, input, classes):
        super(simple_predict, self).__init__()
        
        self.main = nn.Sequential(
            nn.Linear(input, 128),
            nn.ReLU(),
            nn.Dropout(.25),
            
            nn.Linear(128, 64),
            nn.ReLU(),
            nn.Dropout(.25),
            
            nn.Linear(64, len(classes)),
            nn.Softmax(dim=1)
        )
        
        self.classes = classes
    
    def forward(self, wine):
        predict = torch.exp(self.main(wine))
        predicts = torch.tensor([self.classes[x.argmax()] for x in predict],
                                dtype=torch.float, requires_grad=True)
        return predicts
```

```python
model = simple_predict(x.shape[1], classes)
loss_func = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=.005)
```
