```python
## After the analysis, 3 independent variables can be distinguished:
## 1)free sulfur dioxide
## 2)total sulfur dioxide
## 3)alcohol
## These variables were chosen because they have a large scatter in relation to the attribute “quality”
```

```python
##Let's see how these attributes affect the variable "quality"
```

```python
df.groupby('quality')['free sulfur dioxide'].mean().plot(kind="line")
plt.ylabel('free sulfur dioxide')
```

```python
df.groupby('quality')['total sulfur dioxide'].mean().plot(kind='line')
plt.ylabel('total sulfur dioxide')
```

```python
df.groupby('quality')['alcohol'].mean().plot(kind='line')
plt.ylabel('alcohol')
```

```python
## as you can see, the higher the alcohol content, the greater the quality, but not for the indicator of “quality” = 4,
##probably because such wine is cheaper and stronger than “quality” = 3 and 5
##indicators "free sulfur dioxide" and "total sulfur dioxide" for better quality, as you can see,
##should be in the interval from 13 to 14, and about 35, respectively
```

```python
#Let's also check the effect of other variables on the target
```

```python
plt.figure(figsize=(12,12))
plt.subplot(3,4,1)
df.groupby('quality')['fixed acidity'].mean().plot(kind='line')
plt.ylabel('fixed acidity')
plt.subplot(3,4,2)
df.groupby('quality')['volatile acidity'].mean().plot(kind='line')
plt.ylabel('volatile acidity')
plt.subplot(3,4,3)
df.groupby('quality')['citric acid'].mean().plot(kind='line')
plt.ylabel('citric acid')
plt.subplot(3,4,4)
df.groupby('quality')['residual sugar'].mean().plot(kind='line')
plt.ylabel('residual sugar')
plt.subplot(3,4,5)
df.groupby('quality')['chlorides'].mean().plot(kind='line')
plt.ylabel('chlorides')
plt.subplot(3,4,6)
df.groupby('quality')['density'].mean().plot(kind='line')
plt.ylabel('density')
plt.subplot(3,4,7)
df.groupby('quality')['pH'].mean().plot(kind='line')
plt.ylabel('pH')
plt.subplot(3,4,8)
df.groupby('quality')['sulphates'].mean().plot(kind='line')
plt.ylabel('sulphates')
plt.subplots_adjust(wspace=1, hspace=1)
```

```python
#From the graphs, we can conclude that:
#1) the more "sulphates" and "citric acid" the higher the indicator of "quality"
#2) the less "chlorides", "volatile acidity", "density", "pH" the higher the indicator of "quality"
```
