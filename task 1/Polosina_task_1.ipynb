```python
#import library

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#read a file

ds=pd.read_csv('DataSet_NY.csv')
```

```python
#size

print(ds.shape)
```

```python
#info about dataset

print(ds.info())
```

```python
#check values on null

ds.isnull().sum()
```

```python
#delete empty cells

ds = ds.dropna()
print(ds.shape)
```

```python
ds.isnull().sum()
```

```python
ds
```

```python
#main characteristic

ds.describe()
```

```python
#heat map (correlation table)

fig = plt.gcf() 
figsize = fig.get_size_inches()
fig.set_size_inches(figsize * 4)
matrix = np.triu(ds.corr())
sns.heatmap(ds.corr(), annot=True, mask=matrix)
plt.show()
```

We see, that price dont have strong connection with factors.

```python
#histograms of numerical characteristics

ds.hist(edgecolor="black", linewidth=1.2, figsize=(30, 30))
plt.show()
```

```python
plt.figure(figsize=(30, 30))
sns.pairplot(ds, height=3, diag_kind="hist")
plt.show()
```

```python
sns.catplot("neighbourhood_group", data=ds, kind="count", height=8)
plt.show()
```

```python
sns.catplot("room_type", data=ds, kind="count", height=8)
plt.show()
```
